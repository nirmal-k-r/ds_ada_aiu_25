```python
def extract_news(base_url,start,end):
    all_titles=[]
    all_links=[]
    for i in range(start,end):
        print("Extracting page ",i)
        link=base_url + "?page=" + str(i)
        page=get_page(link)
        titles,links=extract_content(page)
        all_titles.extend(titles)
        all_links.extend(links)
    
    save_to_csv(all_titles,all_links,"defi_news.csv")
    print("Total news extracted: ",len(all_titles))


def get_page(link):
    response=requests.get(link)
    page=bs4.BeautifulSoup(response.text,'html.parser')
    return page

def extract_content(page):
    titles=[]
    links=[]
    original_titles=page.select('.sup-title')
    # print(titles)
    for i,title in enumerate(original_titles):
        # print(i,": ",title.getText())
        # link=title.parent
        # print("Link: ",link['href'])
        titles.append(title.getText())
        links.append("https://defimedia.info/" + title.parent['href'])

    return titles,links

def save_to_csv(titles,links,filename):
    df=pd.DataFrame()
    df['Title']=titles
    df['Link']=links
    df.to_csv(filename,index=False,encoding='utf-8')

#main program
# page=get_page("https://defimedia.info/categorie/live-news")
# titles,links=extract_content(page)
# print(titles)
# print(links)

# save_to_csv(titles,links,"test.csv")

base_url="https://defimedia.info/categorie/live-news"
start=0
end=50
extract_news(base_url,start,end)
```

```python
#parallelise requests using multiprocessing using Pool and threading

from multiprocessing import Pool
import threading

def extract_news_parallel(base_url,start,end):
    all_titles=[]
    all_links=[]
    threads=[]

    def worker(link):
        page=get_page(link)
        titles,links=extract_content(page)
        all_titles.extend(titles)
        all_links.extend(links)

    for i in range(start,end):
        print("Extracting page ",i)
        link=base_url + "?page=" + str(i)
        t=threading.Thread(target=worker,args=(link,))
        threads.append(t)
        t.start()
    
    for t in threads:
        t.join()
    
    save_to_csv(all_titles,all_links,"defi_news_parallel.csv")
    print("Total news extracted: ",len(all_titles))

extract_news_parallel(base_url,start,end)
base_url="https://defimedia.info/categorie/live-news"
start=0
end=1000
```

```python
#threadpool
from multiprocessing.pool import ThreadPool

def extract_news_threadpool(base_url,start,end):
    all_titles=[]
    all_links=[]

    def worker(link):
        page=get_page(link)
        titles,links=extract_content(page)
        return titles,links

    links=[]
    for i in range(start,end):
        print("Extracting page ",i)
        link=base_url + "?page=" + str(i)
        links.append(link)
    
    with ThreadPool(50) as pool:
        results=pool.map(worker,links)
    
    for titles,links in results:
        all_titles.extend(titles)
        all_links.extend(links)
    
    save_to_csv(all_titles,all_links,"defi_news_threadpool.csv")
    print("Total news extracted: ",len(all_titles))

base_url="https://defimedia.info/categorie/live-news"
start=0
end=1000
extract_news_threadpool(base_url,start,end)
    
```
