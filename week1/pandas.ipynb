```python
df2.dropna(inplace=True)
df2.shape
```

```python
#remove Other from Gender
#condition based or filter based

df2=df2[df2['Gender']!='Other']
df2.info()
```

```python
df2.head()
```

# encoding
 - map using dictionary
 - LabelEncoder

```python
#map
df2['Gender'].unique()
```

```python
df2['Gender']=df2['Gender'].map({'Male': 0, 'Female': 1})
```

```python
df2.head()
```

```python
#label encoder
from sklearn.preprocessing import LabelEncoder

sleep_quality_enc=LabelEncoder()
sleep_quality_enc.fit(df2['Sleep_Quality'])

print(sleep_quality_enc.classes_)
# print(sleep_quality_enc.inverse_transform(sleep_quality_enc.classes_))

df2['Sleep_Quality']= sleep_quality_enc.fit_transform(df2['Sleep_Quality'])

df2.head()
```

```python
#handle duplicate values
df2.duplicated().sum()
df2.drop_duplicates(inplace=True)
df2.duplicated().sum()
```

```python
# Create age groups (bins)
bins = [0, 18, 30, 45, 60, 100]
labels = ['0-17', '18-29', '30-44', '45-59', '60+']

df3=df.copy()
df3['Age Group'] = pd.cut(df3['Age'], bins=bins, labels=labels)

df3['Age Group'].value_counts()
```

```python
#distribution of male vs female un different age groups

df3.head()
```

```python
categories=df3.groupby(['Age Group', 'Gender']).size().unstack()

categories
```

```python
categories.plot(kind='bar', stacked=True, figsize=(10, 6))
```

Distribution of males vs females is balanced among different age groups. Highest numbers of workers betwee 18 and 60.
