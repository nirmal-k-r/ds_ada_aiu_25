```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
df=pd.read_csv('titatnic.csv')

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df2=df[['Age', 'Fare', 'Sex', 'sibsp','Pclass','Embarked', '2urvived']]

df2.rename(columns={'2urvived':'Survived'}, inplace=True)   

df2.head()
```

```python
df2.isnull().sum()
df2.dropna(inplace=True)
df2.isnull().sum()
```

```python
df2['Survived'].value_counts()
```

```python
y=df2['Survived']
x=df2.drop('Survived', axis=1)
```

```python
#split dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
y_train.value_counts()
```

```python
#logistics regression
#random forest
#naives bayes
```

```python
#logistic regression model
from pyexpat import model
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression()

lr_model.fit(x_train, y_train) #training step

y_pred=lr_model.predict(x_test)

#evaluation

accuracy=accuracy_score(y_test, y_pred)
confusion=confusion_matrix(y_test, y_pred)
report=classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Confusion Matrix:\n", confusion)
print("Classification Report:\n", report)
```

```python
#naive bayes
from sklearn.naive_bayes import GaussianNB
nb_model=GaussianNB()   

nb_model.fit(x_train, y_train) #training step

y_pred_nb=nb_model.predict(x_test)

#evaluation
accuracy_nb=accuracy_score(y_test, y_pred_nb)
confusion_nb=confusion_matrix(y_test, y_pred_nb)
report_nb=classification_report(y_test, y_pred_nb)
print("Accuracy:", accuracy_nb)
print("Confusion Matrix:\n", confusion_nb)
print("Classification Report:\n", report_nb)

```

```python
#random forest model
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train) #training step

y_pred_rf=rf_model.predict(x_test)

#evaluation
accuracy_rf=accuracy_score(y_test, y_pred_rf)
confusion_rf=confusion_matrix(y_test, y_pred_rf)
report_rf=classification_report(y_test, y_pred_rf)

print("Accuracy:", accuracy_rf)
print("Confusion Matrix:\n", confusion_rf)
print("Classification Report:\n", report_rf)
```

```python
#grid tune random forest model
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [10, 20, 40],
    'max_depth': [3,4,5,6,7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 3, 4],
}

#num combinations
param_grid['n_estimators'].__len__()*param_grid['max_depth'].__len__()*param_grid['min_samples_split'].__len__()*param_grid['min_samples_leaf'].__len__()

```

```python
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(x,y)
best_params = grid_search.best_params_
```

```python
best_params 
```

```python
models=grid_search.best_estimator_ #best model after grid search

grid_search.best_score_
```

```python
final_model=RandomForestClassifier(max_depth= 6, min_samples_leaf= 3, min_samples_split= 10, n_estimators= 40)
final_model.fit(x_train, y_train)
y_final_pred=final_model.predict(x_test)
accuracy_final=accuracy_score(y_test, y_final_pred)
confusion_final=confusion_matrix(y_test, y_final_pred)
report_final=classification_report(y_test, y_final_pred)
print("Accuracy:", accuracy_final)
print("Confusion Matrix:\n", confusion_final)
print("Classification Report:\n", report_final)
```
