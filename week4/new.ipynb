```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")

print("Path to dataset files:", path)
```

<h1>loading the dataset into a dataframe <h1>

```python
import pandas as pd

df = pd.read_csv(path + "/creditcard.csv")

print(df.head())
```

```python
df.info()
```

```python
df.describe().T
```

we have a dataset that consists of features (independent variables) and we also have the variable of interest (dependent variable). What are the steps for naive/basic predictions.

```python
X=df.drop(columns=['Class','Time'])
y=df['Class']

print(X.info())
print(y.info())
```

```python
y.value_counts()
```

we see that this is a problem imbalanced dataset.

Now we will split our dataset into train test split but with a small tweak. this tweak is called stratify.

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=42)
```

```python
X.shape 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,      
    random_state=42,     # reproducibility
    stratify=y       # preserve class proportions
)
```

```python
X_train_res,y_train_res=rus.fit_resample(X_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression

lg_model = LogisticRegression(solver='liblinear', random_state=0) 

lg_model.fit(X_train_res, y_train_res)
```

```python
y_pred=lg_model.predict(X_test)
```

hyperparameters greatly influence the performance of our models. choosing the right hyperparameters depend directly on the type of data we have. there is a process called hyper parameter optimization that is used to get the right parameters.

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score, ConfusionMatrixDisplay

cm=confusion_matrix(y_test, y_pred)
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))

disp=ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```
