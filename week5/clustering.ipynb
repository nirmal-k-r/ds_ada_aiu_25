```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('IRIS.csv')
df
```

```python
df2=df.drop(columns=['species'])
df2
```

```python
df2.isna().sum()
```

```python
#pca
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_transformed=pca.fit_transform(df2)


#plot pca
plt.figure(figsize=(8,6))
plt.scatter(pca_transformed[:,0],pca_transformed[:,1])
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('PCA of Iris Dataset')
plt.show()
```

```python
#k-means clustering
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3,random_state=42)
kmeans.fit(df2)
labels=kmeans.labels_

#plot clusters
plt.figure(figsize=(8,6))
plt.scatter(pca_transformed[:,0],pca_transformed[:,1],c=labels
,cmap='viridis')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('K-means Clustering of Iris Dataset')
plt.show()
```

```python
labels
```

```python
df2['Cluster']=labels
df2
```

```python
df2['specicies']=df['species']
df2
```

```python
df2['PCA1']=pca_transformed[:,0]
df2['PCA2']=pca_transformed[:,1]
df2
```

```python
sns.scatterplot(data=df2,x='PCA1',y='PCA2',hue='Cluster',style='specicies',s=100)
```

```python
#wcss score plot to find best value of for elbow method
wcss=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i,random_state=42)
    kmeans.fit(df2.drop(columns=['Cluster','specicies','PCA1','PCA2']))
    wcss.append(kmeans.inertia_)

plt.plot(range(1,10),wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
map={'Iris-setosa':1,'Iris-versicolor':2,'Iris-virginica':0}
df2['specicies']=df2['specicies'].map(map)


df2
```

```python
#score accuracy 
from sklearn.metrics import accuracy_score
accuracy_score(df2['specicies'],df2['Cluster'])
```

```python
#dbi and silhouette score
from sklearn.metrics import davies_bouldin_score,silhouette_score
dbi=davies_bouldin_score(df2.drop(columns=['Cluster','specicies','PCA1','PCA2']),df2['Cluster'])
silhouette=silhouette_score(df2.drop(columns=['Cluster','specicies','PCA1','PCA2']),df2['Cluster'])
print('Davies-Bouldin Index:',dbi)
print('Silhouette Score:',silhouette)
```

```python
#agglomerative clustering
from sklearn.cluster import AgglomerativeClustering
agglo=AgglomerativeClustering(n_clusters=3)
agglo_labels=agglo.fit_predict(df2.drop(columns=['Cluster','specicies','PCA1','PCA2']))
agglo_labels

dbi=davies_bouldin_score(df2.drop(columns=['Cluster','specicies','PCA1','PCA2']),agglo_labels)
silhouette=silhouette_score(df2.drop(columns=['Cluster','specicies','PCA1','PCA2']),agglo_labels)
print('Davies-Bouldin Index:',dbi)
print('Silhouette Score:',silhouette)
```

```python
#spectral clustering
from sklearn.cluster import SpectralClustering
spectral=SpectralClustering(n_clusters=3,affinity='nearest_neighbors',random_state=42)
spectral_labels=spectral.fit_predict(df2.drop(columns=['Cluster','specicies','PCA1','PCA2']))
spectral_labels

dbi=davies_bouldin_score(df2.drop(columns=['Cluster','specicies','PCA1','PCA2']),spectral_labels)
silhouette=silhouette_score(df2.drop(columns=['Cluster','specicies','PCA1','PCA2']),spectral_labels)
print('Davies-Bouldin Index:',dbi)
print('Silhouette Score:',silhouette)
```
